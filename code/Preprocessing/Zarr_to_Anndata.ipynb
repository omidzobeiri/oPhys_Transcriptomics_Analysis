```python
import anndata as ad
from functions.data_loading import load_mouse_zarr
from pathlib import Path 
import numpy as np
   
ZARR_DIR = Path('../scratch/multimodal_data')
MOUSE_IDS = [f.stem.split('_')[0] for f in ZARR_DIR.glob('*_multimodal_data.zarr')]
```

```python
remaining_mouse_ids = [
    mid for mid in MOUSE_IDS
    if not (ZARR_DIR / f"{mid}_stim_evoked_mean.h5ad").exists()
]
print(f"Processing {len(remaining_mouse_ids)} mice: {remaining_mouse_ids}")

# ── Create missing h5ad files from zarr ──
for mid in remaining_mouse_ids:
    print(f"Loading mouse {mid} from zarr...")
    adata_z = load_mouse_zarr(mid, zarr_dir=str(ZARR_DIR), include_genes=True)

    X_z = adata_z.X.toarray() if hasattr(adata_z.X, 'toarray') else np.asarray(adata_z.X)
    baseline = adata_z.baseline.toarray() if hasattr(adata_z.baseline, 'toarray') else np.asarray(adata_z.baseline)

    adata_h5 = ad.AnnData(
        X=X_z,
        obs=adata_z.obs.copy(),
        var=adata_z.var.copy()
    )
    adata_h5.layers['baseline'] = baseline

    adata_h5.write_h5ad(ZARR_DIR / f"{mid}_stim_evoked_mean.h5ad")
```
